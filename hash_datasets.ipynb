```python
import numpy as np
import imagehash
from PIL import Image
from datasets import load_dataset
from huggingface_hub import hf_hub_download
from tqdm import tqdm

def compute_hashes(dataset, image_key, repo_id=None):
    hashes = []
    for data in tqdm(dataset, desc="Computing hashes"):
        img = data[image_key]
        if isinstance(img, str):
            if repo_id is not None:
                # Download image from the Hugging Face Hub
                sub_folder, image_id = img.split('/')
                path = hf_hub_download(
                    repo_id=repo_id,
                    filename=image_id,
                    subfolder=sub_folder,
                    repo_type="dataset"
                )
                image = Image.open(path).convert('RGB')
            else:
                image = Image.open(img).convert('RGB')
        elif isinstance(img, Image.Image):
            image = img
        else:
            continue  # Skip if the image cannot be loaded
        hash = imagehash.average_hash(image)  # You can choose other hash functions
        hash_int = int(str(hash), 16)
        hashes.append(hash_int)
    return np.array(hashes, dtype='uint64')
```

```python
import numpy as np
import imagehash
from PIL import Image
from tqdm import tqdm
from datasets import load_dataset
from huggingface_hub import hf_hub_download

def compute_hashes(dataset, image_key, repo_id=None, hash_function='phash', hash_size=16):
    hashes = []
    for data in tqdm(dataset, desc="Computing hashes"):
        img = data[image_key]
        if isinstance(img, str):
            if repo_id is not None:
                sub_folder, image_id = img.split('/')
                path = hf_hub_download(
                    repo_id=repo_id,
                    filename=image_id,
                    subfolder=sub_folder,
                    repo_type="dataset"
                )
                image = Image.open(path).convert('RGB')
            else:
                image = Image.open(img).convert('RGB')
        elif isinstance(img, Image.Image):
            image = img
        else:
            continue  # Skip if the image cannot be loaded

        # Compute the perceptual hash
        if hash_function == 'phash':
            hash = imagehash.phash(image, hash_size=hash_size)
        elif hash_function == 'whash':
            hash = imagehash.whash(image, hash_size=hash_size)
        else:
            raise ValueError("Unsupported hash function")

        hash_int = int(str(hash), 16)
        hashes.append(hash_int)
    return np.array(hashes, dtype='uint64')
```

```python
import numpy as np
import imagehash
from PIL import Image
from datasets import load_dataset, concatenate_datasets
from huggingface_hub import hf_hub_download
from tqdm import tqdm
import faiss

# Parameterize the hash size
HASH_SIZE = 8  # You can adjust this value (e.g., 8, 16, 24, 32)

# Function to preprocess images
def preprocess_image(image):
    # Convert to grayscale
    image = image.convert('L')
    # Resize to a standard size (matching the expected size for hashing)
    image = image.resize((256, 256))
    # Optional: Apply other normalization techniques if needed
    return image

# Function to compute hashes with preprocessing
def compute_hashes(dataset, image_key, repo_id=None, hash_size=HASH_SIZE):
    hashes = []
    for data in tqdm(dataset, desc="Computing hashes"):
        img = data[image_key]
        if isinstance(img, str):
            if repo_id is not None:
                # Download image from the Hugging Face Hub
                sub_folder, image_id = img.split('/')
                path = hf_hub_download(
                    repo_id=repo_id,
                    filename=image_id,
                    subfolder=sub_folder,
                    repo_type="dataset"
                )
                image = Image.open(path).convert('RGB')
            else:
                image = Image.open(img).convert('RGB')
        elif isinstance(img, Image.Image):
            image = img
        else:
            continue  # Skip if the image cannot be loaded

        # Preprocess the image
        image = preprocess_image(image)

        # Compute the perceptual hash (phash)
        hash_obj = imagehash.phash(image, hash_size=hash_size)

        # Convert hash object to integer
        hash_int = int(str(hash_obj), 16)

        hashes.append(hash_int)
    return np.array(hashes)

# Function to convert hashes to binary arrays suitable for FAISS
def hashes_to_binary_array(hashes, hash_size):
    # Determine the number of bits and bytes needed to represent the hash
    num_bits = hash_size * hash_size  # Total bits in the hash
    num_bytes = (num_bits + 7) // 8   # Number of bytes needed

    # Convert hashes to binary representation
    hashes_bin = np.zeros((len(hashes), num_bytes), dtype=np.uint8)
    for i, hash_int in enumerate(hashes):
        # Convert the NumPy integer to Python int
        hash_int_py = int(hash_int)
        # Convert the integer hash to bytes
        hash_bytes = hash_int_py.to_bytes(num_bytes, byteorder='big')
        hashes_bin[i] = np.frombuffer(hash_bytes, dtype=np.uint8)

    return hashes_bin




# Build the FAISS index on the 'mmvp' dataset (pretend training set)
train_dataset_name = 'mmvp'  # Using 'mmvp' as the pretend training set
train_dataset = loaded_datasets[train_dataset_name]
train_image_key = test_dataset_map[train_dataset_name][-1]

# Compute hashes for the training dataset
train_hashes = compute_hashes(
    train_dataset,
    train_image_key,
    hash_size=HASH_SIZE
)

# Convert training hashes to binary arrays
train_hashes_bin = hashes_to_binary_array(train_hashes, HASH_SIZE)

# Determine the total number of bits in the hash
num_bits = HASH_SIZE * HASH_SIZE

# Build the FAISS index
index = faiss.IndexBinaryFlat(num_bits)
index.add(train_hashes_bin)

# Function to perform the search and count overlaps
def count_overlaps(test_hashes_bin, index, threshold):
    # Perform range search
    result = index.range_search(test_hashes_bin, threshold)
    lims, distances, indices = result

    # Count the number of test images with at least one match
    n_matches_per_query = np.diff(lims)
    n_test_images_with_match = np.sum(n_matches_per_query > 0)

    return n_test_images_with_match

# Set the Hamming distance threshold
threshold = 1  # Adjust as needed

# Initialize a dictionary to store the results
results = {}

# Process each test dataset
for test_name, test_dataset in loaded_datasets.items():
    if test_name == train_dataset_name:
        continue  # Skip the training dataset

    print(f"\nProcessing test dataset: {test_name}")
    test_image_key = test_dataset_map[test_name][-1]
    repo_id = None
    if test_name == 'vstar':
        repo_id = "craigwu/vstar_bench"

    # Compute hashes for the test dataset
    test_hashes = compute_hashes(
        test_dataset,
        test_image_key,
        repo_id=repo_id,
        hash_size=HASH_SIZE
    )

    # Convert test hashes to binary arrays
    test_hashes_bin = hashes_to_binary_array(test_hashes, HASH_SIZE)

    # Count overlaps
    n_overlaps = count_overlaps(test_hashes_bin, index, threshold)
    total_images = len(test_hashes)

    # Store the results
    results[test_name] = {
        'overlaps': n_overlaps,
        'total_images': total_images
    }

    print(f"{n_overlaps} out of {total_images} images overlap with the training set.")

# Tabulate and print the results
print("\nSummary of Overlaps:")
for test_name, data in results.items():
    overlaps = data['overlaps']
    total_images = data['total_images']
    print(f"{test_name}: {overlaps}/{total_images} images overlap with the training set.")
```
